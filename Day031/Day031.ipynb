### Question 3:
Part 2. Create a bar chart that displays the total number of units sold per month. Make sure you have the appropriate labels for the x and y axis.

```python
# Stacked bar graph to demonstrate sales by flavor
# Vanilla	Chocolate	Strawberry	Neapolitan	Mango	Lime
# Compute the bottoms for each subsequent flavor.
b_str = list(np.add(df_monthly_sales['Vanilla'], df_monthly_sales['Chocolate']))
b_neo = list(np.add(b_str, df_monthly_sales['Strawberry']))
b_man = list(np.add(b_neo, df_monthly_sales['Neapolitan']))
b_lim = list(np.add(b_man, df_monthly_sales['Mango']))

# plot the bar graphs...
plt.bar(df_monthly_sales['Month'], df_monthly_sales['Vanilla'], label="Vanilla")
plt.bar(df_monthly_sales['Month'], df_monthly_sales['Chocolate'], bottom=df_monthly_sales['Vanilla'], label="Chocolate")
plt.bar(df_monthly_sales['Month'], df_monthly_sales['Strawberry'], bottom=b_str, label="Strawberry")
plt.bar(df_monthly_sales['Month'], df_monthly_sales['Neapolitan'], bottom=b_neo, label="Neapolitan")
plt.bar(df_monthly_sales['Month'], df_monthly_sales['Mango'], bottom=b_man, label="Mango")
plt.bar(df_monthly_sales['Month'], df_monthly_sales['Lime'], bottom=b_lim, label="Lime")
plt.title("Profits across all months by flavor") #title
plt.xlabel("Months") #x label
plt.ylabel("Sales in dollars") #y label
plt.legend()
```
